Housing Data subjected to Regression. The key metrics will be validated and the explanation done for each parameter.

```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

```
housingdf = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Data/home_data.csv')

housingdf.head(5)
```

```
#Use Statsmodels to run linear regression

X = housingdf[['bedrooms','bathrooms','sqft_living','sqft_basement','sqft_living15','sqft_lot','sqft_above','yr_built','grade','floors']]
Y = housingdf['price']

X = sm.add_constant(X) # adding a constant
model = sm.OLS(Y, X).fit()
predictions = model.predict(X) 

print_model = model.summary()
print(print_model)

columns = ['bedrooms','bathrooms','sqft_living','sqft_basement','sqft_living15','sqft_lot','sqft_above','yr_built','grade','floors']
hx = housingdf[columns]
hy = housingdf['price']

x_train,x_test,y_train,y_test = train_test_split(hx,hy,train_size = 0.7,random_state=42)
x_train_new = sm.add_constant(x_train)
x_test_new = sm.add_constant(x_test)
print ("\nVariance Inflation Factor")
cnames = x_train.columns
for i in np.arange(0,len(cnames)):
    xvars = list(cnames)
    yvar = xvars.pop(i)
    mod = sm.OLS(x_train[yvar],sm.add_constant(x_train_new[xvars]))
    res = mod.fit()
    vif = 1/(1-res.rsquared)
    print (yvar,round(vif,3))
```

Interpreting the results above:
1. The R2 is a +ve number and its 0.6 which is less than 0.7, hence the model is not fit enough for selection
2. The VIF or variance inflation factor for sqft_living, sqft_basement and sqft_above is showing inf which is due to R2 being 1 - meaning perfect fit
3. The Durbin-Watson is less than 2 which tests the null hypothesis on the auto-correlation of residuals and this range indicates +ve correlation

```
winedf = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Data/winequality-white.csv',sep=';')

winedf.head(5)
```

Iteration-1

```
#Use Statsmodels to run linear regression

X = winedf[['fixed acidity','volatile acidity','citric acid','residual sugar','chlorides','free sulfur dioxide','total sulfur dioxide','density','pH','sulphates','alcohol']]
Y = winedf['quality']

colnms = ['volatile acidity',  'chlorides', 'free sulfur dioxide','total sulfur dioxide','pH', 'sulphates', 'alcohol']

pdx = winedf[colnms]
pdy = winedf["quality"]

x_train,x_test,y_train,y_test = train_test_split(pdx,pdy,train_size = 0.7,random_state=42)
x_train_new = sm.add_constant(x_train)
x_test_new = sm.add_constant(x_test)

full_mod = sm.OLS(y_train,x_train_new)
full_res = full_mod.fit()
print ("\n \n",full_res.summary())

print ("\nVariance Inflation Factor")
cnames = x_train.columns
for i in np.arange(0,len(cnames)):
    xvars = list(cnames)
    yvar = xvars.pop(i)
    mod = sm.OLS(x_train[yvar],sm.add_constant(x_train_new[xvars]))
    res = mod.fit()
    vif = 1/(1-res.rsquared)
    print (yvar,round(vif,3))
```

1. The R2 value is very low - 0.26 which is far from 0.7 for selecting as fit
2. The p-value of 'total sulfur dioxide' and 'chlorides' is more 0.05 which means that these variables are insignificant in the regression analysis
3. The Variance inflation factor of 'total sulfur dioxide' is the highest which influences the residual error negatively.
4. In the next iteration, we will remove total sulfur dioxide and chlorides for model optimization.
5. Also, the Durbin-Watson number is slightly more than 2 which indicates that the residuals are negatively correlated
